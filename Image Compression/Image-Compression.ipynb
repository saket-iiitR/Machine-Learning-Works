```python
#here we are going to Quantize image with the help of OpenCV KMean Clusturing
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img=cv2.imread("Images\landscape.jpg",1) 
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
Z=img.reshape((-1,1))
#Samples must be of np.float32 datatype,
#and each feature should be put in a single column.
Z=np.float32(Z)
Z
```

```python
# Criteria:-It is the iteration termination criteria.
# When this criteria is satisfied, algorithm iteration stops.
# Actually, it should be a tuple of 3 parameters.
```

```python
criteria= (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER,10,1.0)
```

```python
K=2
ret,label1,center1=cv2.kmeans(Z,K,None,criteria,10,cv2.KMEANS_RANDOM_CENTERS)
```

```python
center1=np.uint8(center1)
res1=center1[label1.flatten()]
output1=res1.reshape((img.shape))
```

```python
plt.imshow(output1)
```
