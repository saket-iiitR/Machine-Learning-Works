# Data Restructure

```python
weather=pd.read_csv('data/data/weather.csv')
weather.head()
```

```python
weather_melt=pd.melt(weather, 
          id_vars=['id','year','month','element'],
          var_name='day',
         value_name='temp')
weather_melt.head()
```

```python
weather_tidy=weather_melt.pivot_table(index=['year','month','day','id'],
                        columns='element',
                        values='temp',
                        dropna=False).reset_index()
weather_tidy.head()
```

```python
weather_tidy.to_csv('output/weather.csv') 
#if we want to save the datadet as tab separated dat
#weather_tidy.to_tsv('output/weather.tsv', sep=''/t') 
```

```python
weather_tidy.loc[(weather_tidy.month==1) & (weather_tidy.day=='d15'),['tmax']]
```
