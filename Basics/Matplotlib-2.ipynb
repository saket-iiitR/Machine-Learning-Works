## Multiple curves in the same plot

```python
x = np.arange(-10,20)
y = x ** 2
plt.plot(x,y,'c',label='square', linewidth=2)

x2 = np.arange(-10,20)
y2 = x2 + 30
plt.plot(x2,y2,'m',label='linear',linewidth=4)

plt.legend()

plt.text(14, 190, 'Hello')

plt.grid(True,color='k')

plt.show()
```

```python
x = np.arange(-10,20)
y = x ** 2
plt.plot(x,y,'b^', x, y, 'r--')

plt.legend()

plt.show()
```

# Vectorize / Apply

```python
x = np.arange(1,101)
y = np.log(x)
plt.plot(x, y)
plt.show()
```

```python
import math
x = np.arange(1,101)
y = np.vectorize(math.log)(x)
plt.plot(x, y)
plt.show()
```

```python
import math
x = np.arange(1,101)
y = [ math.log(i) for i in x ]
plt.plot(x, y)
plt.show()
```

```python
import math
x = np.arange(1,101)
y = list(map(math.log, x))
plt.plot(x, y)
plt.show()
```

## Subplots

```python
x = [1, 2, 3, 4, 5, 6]
y = [1, 2, 3, 4, 5, 6]

plt.subplot(221) # rows, cols, plot-number
plt.plot(x, y, 'r-')

plt.subplot(222)
plt.plot(x, y, 'g--')

plt.subplot(2, 2, 3)
plt.plot(x, y, 'bo')
plt.plot(x-2, y+2, 'bo')

plt.subplot(2, 2, 4)
plt.plot(x, y, 'y^')

plt.show()
```

## Distributions

### From numpy documentation

```python
mu, sigma = 0, 0.1 # mean and standard deviation
s = np.random.normal(mu, sigma, 100)
s
```

```python
# Verify mean
abs(mu - np.mean(s))
```

```python
# Verify standard deviation
abs(sigma - np.std(s))
```

```python
count, bins, ignored = plt.hist(s, 30, normed=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2) ), linewidth=2, color='r')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
n, bins, patches = plt.hist(x, 50, normed=1, facecolor='b', alpha=0.5)



plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```

```python
import math

sigmoid = lambda i : 1 / (1 + math.exp(-i))

x = np.arange(-10., 10., 0.2)
y = [sigmoid(i) for i in x]

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

# Move both axes
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')

# Hide unwanted axes
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# Show ticks in left and bottom axes only
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

plt.plot(x,y)
plt.show()
```
