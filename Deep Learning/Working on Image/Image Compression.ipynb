```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize']=[15,10]
```

```python
im=cv2.imread('Self/Images/landscape.jpg')
image=cv2.cvtColor(im,cv2.COLOR_BGR2RGB)
plt.imshow(np.hstack([im,image]))
```

```python
(h,w,c)=image.shape
(h,w,c)
```

```python
from sklearn.cluster import MiniBatchKMeans
```

```python
image=image.reshape(h*w, 3)
```

```python
numClusters=16
```

```python
clt=MiniBatchKMeans(n_clusters=numClusters)
labels=clt.fit_predict(image)
```

```python
labels.shape #labes me centres honge na ki va;lue
```

```python
clt.cluster_centers_.astype('uint8') #returns the centres
```

```python
quant=clt.cluster_centers_.astype('uint8')[labels]
```

```python
quant=quant.reshape((h,w,3))
```

```python
plt.imshow(quant)
plt.show()
```

```python
quant= cv2.cvtColor(quant,cv2.COLOR_RGB2BGR)
cv2.imwrite('self/compressed.jpg',quant)
```
